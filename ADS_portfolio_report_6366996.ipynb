#### short literature review
  Traditionally, well-being was classified into happiness and life satisfaction. Well-being has been defined as the combinaton of feeling good and functioning well; the experience of positive emotions such as contentment and happiness, and the development of one's potential, having some control over one's life, having a sense of purpose, and experiencing positive relationships (Huppert, 2009). 

Multidimentional measurements were used in the current survey to indicate the well-being situation of Dutch      participants. Five aspects of satisfactions were included, respectively satisfaction with education and work, 
satisfaction with health situation, satisfaction with finance, satisfaction with living environment, satisfaction with social life.  A composite measure of many underlying dimensions could bring more insights into well-being than just a single item measured by happiness or life satisfaction. These two aspects could represent the general pattern of well-being but cannot show which aspects matter for different level of well-being. More detailed measurement could make it easier for policymakers or individuals to target interventions. Demographic categories such as gender, age, education level, migration background were recorded in current dataset. Previous study have shown that the older and retired populations showed lower life satisfaction or happiness (Ruggeri et al., 2020). Such finding is in line with other studies that showed ageing beyong 75 is often associated with increased loneliness and isolation (Nicolaisen et al., 2014; Steptoe et al., 2013). Apart from age, education level and migration background may have relationship with well-being. Investigating their relationship with well-being can give instructions to target specific groups of people in order to improve their well-being.

```python
# linear regression between education level and life satisfaction score (ScoreLifeSatisfaction_5)
from scipy.stats import shapiro 
from sklearn.linear_model import LinearRegression

# create dummy variables for the categorical variable "Education level"
X = pd.get_dummies(data=df_2['PersonCharacteristics'])
```

```python
print(X)
```

```python
# test assumption of normality of the predicted variable
nor_life_satisfaction = shapiro(df_2['ScoreLifeSatisfaction_5'])
print(nor_life_satisfaction)
```

Unluckily, the normality assumption of the predicted variable (score of general life satisfaction) was violated. The reason could be that these scores are already averaged scores instead of the raw score for each individual, which cannot represent the sample distribution anyway. 

Actually, the dependent and independent variables in a regression model do not need to be normally distributed by themselves--only the prediction errors need to be normally distributed. Let's do the linear regression to have a look how will the data be fitted.

```python
# do regression 
regr = LinearRegression()
regr.fit(X, df_2['ScoreLifeSatisfaction_5'])
predicted = regr.predict(X)
print(predicted)

# get parameters of the regression model, coefficients and R-square
coefficients = regr.coef_
rsquare = regr.score(X, df_2['ScoreLifeSatisfaction_5'])

print(f"The coefficients of the regression are : {coefficients}")
print(f"The R square statistic is : {rsquare}")
```

```python
# make a plot for the regression model
plt.scatter(df_2["PersonCharacteristics"], df_2['ScoreLifeSatisfaction_5'], color='black')
plt.plot(df_2["PersonCharacteristics"], predicted, color='blue',linewidth=3)
plt.xlabel("Education level")
plt.ylabel("Score of general life satisfaction")
plt.show()
```

#### results
The current results of statistical analysis suggested that there is a general fit between education level and the score of general life satisfaction, the R-square statistic is 0.817, predicting that around 81.7% of general life satisfaction can be explained by education level. However, it should be noted that here we have not include other predictors yet, so whether such relationship is the true situation, we cannot make conclusion about it now. It can be only predicted that higher education level does correlate with higher general life satisfaction score, not a causation but a correlation.

```python
# start descriptive analysis, t-test assumption check and t-test    
from scipy.stats import levene

# describe datasets between groups

print("For people work more than 12 hours/week:", "\n", df_3_work['ScoreHappiness_1'].describe())
print("For people work less than 12 hours/week:", "\n", df_3_no_work['ScoreHappiness_1'].describe())
```

```python
# Levene's test for homogeneity of variances between groups 
levene_test = levene(df_3_work['ScoreHappiness_1'],df_3_no_work['ScoreHappiness_1']) 
print(levene_test)
```

As the result of Levene's test suggested, the two cluster of data conform to the variance criterion (p > 0.05), in other words, the variance between the two groups of data is equal.

```python
# test assumption of normality 
nor_work = shapiro(df_3_work['ScoreHappiness_1']) 
nor_no_work = shapiro(df_3_no_work['ScoreHappiness_1']) 

print("For people work more than 12 hours/week:", "\n", nor_work)
print("For people work less than 12 hours/week:", "\n", nor_no_work)
```

As the normality test suggested, the two groups of data violated the assumption of normality (p < 0.05). Concerning the results of Levene's test and normality test, Mann-Whitney U test should be used in order to compare the means of two groups of data (people work more than 12 hours/week v.s. people work less than 12 hours/week).

```python
# do Mann-Whitney U test to compare the two groups of data
from scipy.stats import mannwhitneyu

stat, p = mannwhitneyu(df_3_work['ScoreHappiness_1'], df_3_no_work['ScoreHappiness_1'])
print(f"The statistic is {stat: .3f}", "\n",
      f"p-value = {p: .3f}")

alpha = 0.05
if p > alpha:
	print("There is no significant difference in general happiness level between people work more or less than 12 hours/week.")
else:
	print("People work less than 12 hours/week feel significantly happier than people work more than 12 hours/week")
```

#### Results
As can be seen from the statistical testing, there is no significant difference in general happiness level between people who work more or less than 12 hours/week. Even though there reported more high scores in general happiness within people who work less, such difference did not reach a significant population level.

```python
# plot the general well-being scores in three groups of people
concatenated = pd.concat([df_4_dutch.assign(dataset='df_4_dutch'),df_4_western.assign(dataset='df_4_western'),df_4_no_western.assign(dataset='df_4_no_western')], sort=True)

ax = sns.boxplot(x='dataset', y='well_being', data=concatenated)
ax.set_xlabel('Migration background')
ax.set_ylabel('Score of general well-being')
print(ax)
```

It is obvious from the boxplots above that people with Dutch background stay in a relatively high level of well-being. Meanwhile, people with western background are quite similar with Dutch, but in a broader range of scores. For people with non-western background, apparently they scored lower compared the other two groups of people. Whether such difference is significant should be tested in further statistical analysis. 

```python
# describe datasets among groups

print("For people with a Dutch background:", "\n", df_4_dutch['well_being'].describe())
print("For people with a Western background:", "\n", df_4_western['well_being'].describe())
print("For people with a Non-western background:", "\n", df_4_no_western['well_being'].describe())
```

```python
# Levene's test for homogeneity of variances among groups 
levene_test = levene(df_4_dutch['well_being'],df_4_western['well_being'],df_4_no_western['well_being']) 
print(levene_test)
```

As the result of Levene's test suggested, the three groups of data conform to the variance criterion (p > 0.05), in other words, the variance among the three groups of data is equal.

```python
# test assumption of normality 
nor_dutch = shapiro(df_4_dutch['well_being']) 
nor_western = shapiro(df_4_western['well_being'])
nor_no_western = shapiro(df_4_no_western['well_being'])

print("For people with a Dutch background:", "\n", nor_dutch)
print("For people with a Western background:", "\n", nor_western)
print("For people with a Non-western background:", "\n", nor_no_western)
```

As the results of normality test showed, the data of people with Dutch background violated the assumption of normality (p < 0.05) while the other two groups of data conform to the normal distribution. It is difficult to obtain a normal distribution for each group of data here, since in every group there are only seven values, suggesting small samples of data. It can be predicted from the boxplots that the distribution of the Dutch group has a positive skew (many large values). Let's transform the data and do one-way ANOVA afterwards.

```python
# perform one-way ANOVA 
from scipy.stats import f_oneway

group1 = np.log(df_4_dutch['well_being']) #log-transform the original data in order to make big values near the center of distribution
group2 = np.log(df_4_western['well_being'])
group3 = np.log(df_4_no_western['well_being'])

F, p = f_oneway(group1, group2, group3)
print(f"The F-value is {F: .3f}", "\n",
      f"p-value = {p: .3f}")

alpha = 0.05
if p > alpha:
	print("There is no significant difference in general well-being level among people with three different migration background.")
else:
	print("There is at least one group of people differ significantly from the rest in general well-being level")
```

```python
# post-hoc tests (pairwise comparison)
# Tukey's HSD (honest significant difference) test will be used here

import statsmodels.stats.multicomp as mc

# preformat data in Group and value columns
df_concatenated = pd.concat([df_4_dutch.assign(dataset='Dutch'),df_4_western.assign(dataset='Western'),df_4_no_western.assign(dataset='Non-western')])

comp = mc.MultiComparison(pd.to_numeric(df_concatenated['well_being'], errors='coerce'), df_concatenated['dataset'])
post_hoc_res = comp.tukeyhsd(0.05)
post_hoc_res.summary()
```

As can be seen from the results of Tukey's HSD test, the difference of general well-being level between people with Dutch background and people with non-western background, and between western background and non-western background are both significant (p-adjusted = 0.001, < 0.05). Let's plot the difference below: 

```python
post_hoc_res.plot_simultaneous(ylabel= "Migration background", xlabel= "Score of general well-being")[0] #it will produce two same figures without [0]
```

#### Results
It can be predicted from the statistical analyses that migration background has relationship with the overall level of people's well-being. Specifically, people with Dutch and western background showed significantly higher level of well-being in the Netherlands compared to people with non-western background.

```python
# post-hoc tests (Games-Howell test when variances among groups are unequal)
from pingouin import pairwise_gameshowell

posthoc_edu = pairwise_gameshowell(data=df_education_work, dv='Education_work', between='dataset').round(3)

print(posthoc_edu)
```

The results of normality test showed only the data of people between 35 and 44 years old violated the assumption of normality. Since the variance among groups are equal, let's transform the data and do one-way ANOVA afterwards.

```python
# perform one-way ANOVA 

g1 = np.log(df_health['Health'][df_health['dataset'] == '18_24_years_old'])
g2 = np.log(df_health['Health'][df_health['dataset'] == '25_34_years_old'])
g3 = np.log(df_health['Health'][df_health['dataset'] == '35_44_years_old'])
g4 = np.log(df_health['Health'][df_health['dataset'] == '45_54_years_old'])
g5 = np.log(df_health['Health'][df_health['dataset'] == '55_64_years_old'])
g6 = np.log(df_health['Health'][df_health['dataset'] == '65_74_years_old'])
g7 = np.log(df_health['Health'][df_health['dataset'] == '75_older_years_old'])

F, p = f_oneway(g1, g2, g3, g4, g5, g6, g7)
print(f"The F-value is {F: .3f}", "\n",
      f"p-value = {p: .3f}")

alpha = 0.05
if p > alpha:
	print("There is no significant difference in satisfaction with health among people of different age.")
else:
	print("There is at least one group of people differ significantly from the rest in satisfaction with health")
```

```python
# post-hoc tests (pairwise comparison)
# Tukey's HSD (honest significant difference) test 

comp = mc.MultiComparison(df_health['Health'], df_health['dataset'])
post_hoc_res = comp.tukeyhsd(0.05)
post_hoc_res.summary()
```

According to the one-way ANOVA and post-hoc test afterwards, there are many significant difference between different age groups in satisfaction with health. Firstly, young people between 18-24 years old scored higher than any other groups in satisfaction with health. This is in line with the common thought that young people feel healthier due to young body. Secondly, people between 25-34 years old scored higher than other older age groups except people between 65-74 years old. Thirdly, people between 35-44 years old did not show higher satisfaction with health when compared to older age group. This finding is interesting, suggesting 35-year-old as a special age to "feel healthy or not any more".
